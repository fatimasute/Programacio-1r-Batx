Respresentació de paràboles. Aquest és el programa bàsic.

```python
import numpy as np
from matplotlib import pyplot as plt

a = int(input('Introdueix un paràmetre a per la funció y = ax^2 + bx + c : '))
b = int(input('Introdueix un paràmetre b per la funció y = ax^2 + bx + c : '))
c = int(input('Introdueix un paràmetre c per la funció y = ax^2 + bx + c :'))

#Calcular les x màxima i mínima; quantes x es volen i càlcul de les y.
xmax = -4*b/a  
xmin = 2*b/a   

x = np.linspace(xmin, xmax, 1399)
y = a*(x**2) + b*x + c

#Customització de la gràfica.
plt.axhline(y=0, color = 'black', linestyle="-")
plt.axvline(x=0, color = 'black')
plt.xlim(min(xmin, xmax), max(xmin, xmax))
plt.xlabel('Eix x', color = 'orange', fontsize = '22')
plt.ylabel('Eix y', color = 'orange', fontsize = '22')
tit = '\ny = '+str(a)+" x^2 + "+str(b)+ " x + " + str(c)+ '\n'
plt.title(tit, fontsize = 25, color = 'purple')


plt.plot(x, y, color = 'red', linestyle="dotted")
plt.show()
```
