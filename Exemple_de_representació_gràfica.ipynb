Estudia el següent codi, que serveix per representar una llista de dades. Prova de canviar els diferents paràmetres i investiga per a què serveixen cadascun d'ells

```python
import numpy as np                   # Importa el paquet numpy per treballar amb funcions matemàtiques
from matplotlib import pyplot as plt # Importa el paquet pyplot per fer representacions grafiques

ys =4.5+ np.random.rand(50)          # Genera números aleatoris al voltant de 4.5
[x for x in range(len(ys))]          # Genera la llista [0,1,2,...... len(yn)]

plt.figure(figsize=(17, 4))          # Fixa la mida del gràfic
plt.plot(x, ys)                      # Representa els valors de ys respecte a les abcisses x 
plt.fill_between(x, ys, 4, ys > 5, facecolor='g', alpha=0.5)    # Acoloreix les dades superiors a 5, des de 4

plt.title("AQUI POSEM EL TÍTOL QUE VULGUEM")     # Li posa títol al gràfic
plt.show()                                       # Acaba de representar el gràfic
 
```
